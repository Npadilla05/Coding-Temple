```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv(r'C:\Users\Noki\Downloads\AB_NYC_2019 (1).csv')
air_bnb_copy = air_bnb
display(air_bnb_copy)
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb_copy.groupby('neighbourhood_group').count()[['availability_365']]

# There are 5 main neighbourhood groups in New York
# Manhattan is the area that shows up the most and has the highest availability count.
 
```

```python
# Are private rooms the most popular in manhattan?
room_type = air_bnb['room_type'].value_counts().sort_values(ascending=False)
print(room_type)

# There are only 3 room types listed in this data
# Entire home/apt are the most popular found in this data
```

```python
# Which hosts are the busiest and based on their reviews?
busy_host = round(air_bnb.groupby('host_name').sum()).sort_values('number_of_reviews', ascending=False)
display(busy_host)

# We can observe that Michael is the busiest host based on the number of reviews 
```

```python
#Which neighorhood group has the highest average price?
avg_group_price = round(air_bnb.groupby('neighbourhood_group').price.mean(), 2).sort_values(ascending=False)
display(avg_group_price)

# Manhattan, on average is the most expensive area to stay in
```

```python
# Which neighbor hood group has the highest total price?
highest_group_price = round(air_bnb.groupby('neighbourhood_group').price.sum(), 2).sort_values(ascending=False)
display(highest_group_price)

# Manhattan has the most expensive total price of all airbnbs
```

```python
#Which top 5 hosts have the highest total price?
host_price = air_bnb.groupby('host_name').sum().sort_values('price', ascending=False)
display(host_price)

# Airbnb hosts: Sonder, Blueground, Michael, David and Alex are the top 5 hosts with the highest total price
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# First, we filter out the hosts with zero availability
zero_avail_hosts = air_bnb[air_bnb['availability_365'] == 0]
print(zero_avail_hosts)

# Second, we filter for hosts with a review count of 100 or more
review_count_hosts = zero_avail_hosts[zero_avail_hosts['number_of_reviews'] >= 100]
print(review_count_hosts)

# Finally, we print the hosts that meet these requirements
display(review_count_hosts.iloc[:, :])
```

```python
# What host has the highest total of prices and where are they located?
host_prices_location = air_bnb.groupby(['host_id', 'host_name', 'neighbourhood'])['price'].sum()
host_prices_location = host_prices_location.sort_values(ascending=False)
display(host_prices_location)

# Sonder (NYC) is the host with the highest total price and the host is located in the Financial District
```

```python
# When did Danielle from Queens last receive a review?

# Filter to find Danielle from Queens
danielle = air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')]
display(danielle)

# Sort the filtered data by the review date in descending order
sorted_reviews = danielle.sort_values('last_review', ascending=False)
display(sorted_reviews)

# Print the date of the last review for Danielle from Queens
display(sorted_reviews['last_review'].iloc[0])
```

## Further Questions

1. Which host has the most listings?

```python
# Used group by and count to filter thru host names and their listing count
listings_per_host = air_bnb.groupby('host_name')['calculated_host_listings_count'].count()

# Sort the total listings from highest to lowest to find the host with the most listings
listings_per_host = listings_per_host.sort_values(ascending=False)

display(listings_per_host)
```

2. How many listings have completely open availability?

```python
# Used length to filter through the whole data in 'availability_365' to find the number of listings with open availability
open_availability = len(air_bnb[air_bnb['availability_365'] == 365])

display(open_availability)
```

3. What room_types have the highest review numbers?

```python
# Used group by and mean to filter thru room types and find the mean of number of reviews
room_type_review_means = air_bnb.groupby('room_type')['number_of_reviews'].mean()

# Sort the listings from highest to lowest to find which room types have the highest review numbers
room_type_review_means = room_type_review_means.sort_values(ascending=False)

display(room_type_review_means)
```
