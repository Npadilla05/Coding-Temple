### Problem 1:
Check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 5 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat

**BONUS POINTS IF YOU CAN POPULATE THE IMAGE!**

```python
import requests
from bs4 import BeautifulSoup
# URL 
url = "https://pokeapi.co/api/v2/pokemon/"

# List of 5 Pokemon to get data for
pokemon_list = ["bulbasaur", "charmander", "squirtle", "pikachu", "eevee"]

# Loop through each Pokemon in the list and get its data
for pokemon_name in pokemon_list:
    response = requests.get(url + pokemon_name)
    pokemon_data = response.json()

    # Get information from the dictionary
    name = pokemon_data["name"]
    ability_name = pokemon_data["abilities"][0]["ability"]["name"]
    base_experience = pokemon_data["base_experience"]
    front_shiny_url = pokemon_data["sprites"]["front_shiny"]
    attack_base_stat = pokemon_data["stats"][1]["base_stat"]
    hp_base_stat = pokemon_data["stats"][0]["base_stat"]
    defense_base_stat = pokemon_data["stats"][2]["base_stat"]

    # Print out the data for the current Pokemon
    print("Name:", name)
    print("Ability Name:", ability_name)
    print("Base Experience:", base_experience)
    print("Front Shiny URL:", front_shiny_url)
    print("Attack Base Stat:", attack_base_stat)
    print("HP Base Stat:", hp_base_stat)
    print("Defense Base Stat:", defense_base_stat)
    print()
```

### Problem 2:

I want to build the best Pokemon team I can and I need your help to do so! Using the API, build me a team of 6 pokemon who's base-stats are above 60. Return any additional information you think is needed to help understand why you chose that pokemon for the team!

```python
import requests

# URL for the PokeAPI
url = "https://pokeapi.co/api/v2/pokemon/"

# List of Pokemon to choose from. Found these 6 based off the highest base stats i could easily find
pokemon_list = ["ivysaur", "venusaur", "charizard", "blastoise", "pidgeot", "fearow"]

# List of 6 Pokemon to add to the team
team = []

# Loop through the Pokemon list
for pokemon_name in pokemon_list:
    # Send a GET request to the URL for the Pokemon's data
    response = requests.get(url + pokemon_name)
    pokemon_data = response.json()

    # Check if the Pokemon's base stats are all above 60
    base_stats = pokemon_data["stats"]
    if all(stat["base_stat"] > 60 for stat in base_stats):
        # If the Pokemon meets our criteria, add it to the team
        name = pokemon_data["name"]
        types = [type_data["type"]["name"] for type_data in pokemon_data["types"]]
        base_stats_dict = {stat["stat"]["name"]: stat["base_stat"] for stat in base_stats}
        abilities = [ability_data["ability"]["name"] for ability_data in pokemon_data["abilities"]]
        moves = [move_data["move"]["name"] for move_data in pokemon_data["moves"][:4]]

        pokemon = {
            "name": name,
            "types": types,
            "base_stats": base_stats_dict,
            "abilities": abilities,
            "moves": moves
        }
        team.append(pokemon)

for pokemon in team:
    print(pokemon["name"])
    print("Types:", pokemon["types"])
    print("Base Stats:", pokemon["base_stats"])
    print("Abilities:", pokemon["abilities"])
    print("Moves:", pokemon["moves"])
    print()
```

### Problem 3:

Create a function that does the following:
* Input: pokemon_name(user_defined) -> str
* Output: list object containing a dictionary of values from the response object.
* Please include the type, the most current raritey level, at least one held item, and all stats

```python
import requests

def pokemon_data(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}/"

    response = requests.get(url)
    pokemon_data = response.json()

    types = [type_data["type"]["name"] for type_data in pokemon_data["types"]] # Grabs the name of each Pokemon
    held_items = [item_data["item"]["name"] for item_data in pokemon_data["held_items"]] # Grabs the name of each held item of the Pokemon  
    stats = {stat["stat"]["name"]: stat["base_stat"] for stat in pokemon_data["stats"]} # Grabs the name and each stat of the pokemon as a dictionary. Will show attack, defense and speed

    result = [{
        "type": types,
        "held_items": held_items,
        "stats": stats
    }]

    return result

pokemon_data("pikachu")
```
