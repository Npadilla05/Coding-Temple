```python
class ToMongo():
    def __init__(self):
        load_dotenv()
        self.mongo_url = os.getenv('MONGO_URL')
        self.client = pymongo.MongoClient(self.mongo_url)
        self.db = self.client.db
        self.student_perf = self.db.student_perf

    def upload_one_by_one(self):
        ''' Uploads an item to mongoDB one by one '''
        for i in self.db.index: # iterating thru each item inside df index
            self.student_perf.insert_one(self.df.loc[i].to_dict()) # For each item we will insert one or w/e we are iterating thru. iterate thru that using index

if __name__ == '__main__':
    c = ToMongo()
    print(c.upload_one_by_one)
```
