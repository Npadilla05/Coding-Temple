### Convert KM to MI

```python
# miles = km * 0.621371
# Converting km to mi
new = []
for value in df_resort['longest_run']:
    if type(value) == float:
        value = '0.0 mi'
        new.append(value)
    elif type(value) == str:
        l = value.split()
        if l[1] == 'km':
             l[0] = str(round(float(l[0])* 0.621371,1))
             l[1] = 'mi'
             new.append(' '.join(l))
        else:
            new.append(value)

print(new)
df_resort['longest_run'] = new
```
